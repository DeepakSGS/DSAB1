### Activity Description:

    Jio fibre wants to implement AI/ML in their systems.
    The activity task is to build a network intrusion detector,
    a predictive model capable of distinguishing between 'bad' connections,
    called intrusions or attacks, and 'good'; normal connections.
    
    This database contains a standard set of data to be audited, which
    includes a wide variety of intrusions in the network environment.
    Your target is building a machine efficient learning model using
    probabilistic approach models, probabilities of bad or good
    connections using naïve bayes model. This is a multi-class classification.
    Fine tune the model and make a model which has at least 97% accuracy.

### Objective:
    To build a predictive model using Naive Bayes Classifier which is capable of
    distinguishing between 'bad connections: intrusion' and  'good connections: normal'
    
    Note:
    Our ML objective is to classify connections based on feature inputs. Here, it's said
    that it is a multi-class classification problem. Intrusions are classified to several
    classes and we have to sort them out and finally make it to binary classification.
    
    We can go for Naive Bayes Multinomial Classifier for discrete distributions.

## Model Creation:

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
#defining feature'X' & target'y'
X = df.drop(["label"], axis=1)
y = pd.DataFrame(df["label"])
X.head()
```

```python
#Feature Scaling using MinMax Scaler
minmax = MinMaxScaler()
X = minmax.fit_transform(X)
print(X)
```

```python
#Split test and train data 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
#Multinomial Naive Bayes Classifier
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
#fitting model to train set
model.fit(X_train, y_train)
```

```python
#prediction
y_pred = model.predict(X_test)
```

```python
#scores of train and test set
print("Train score is:", model.score(X_train, y_train))
print("Test score is:", model.score(X_test, y_test))
```

    Train and test set scores are above 98%

```python
#calculating accuracy using accuracy_score function
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pred, y_test)
print ("Accuracy is {}.".format(round(accuracy,4)))
```

```python
#evaluating the predicted results: confusion matrix

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score

cmat = confusion_matrix(y_test, y_pred)

#accuracy calculation using formula:
#calculating the total correct predictions 
tot_correct_predictions=0
for i in range(len(cmat)):
    tot_correct_predictions+=cmat[i][i]
    
#calculating the model accuracy mathematically:
accuracy = (tot_correct_predictions / np.sum(cmat))*100

#display accuracy:
print(f'Accuracy obtained on this test set : {accuracy:.2f} %')
```

```python
#Confusion Matrix plot for MultinomialNB Classifier:

cmat = confusion_matrix(y_test, y_pred)
x_axis_labels = ["0", "1"]
y_axis_labels = ["0", "1"]

f, ax = plt.subplots(figsize =(7,7))
sns.heatmap(cmat, annot=True, linewidths=0.2, linecolor="black", fmt=".0f", ax=ax, cmap="vlag", xticklabels=x_axis_labels, yticklabels=y_axis_labels)
plt.xlabel("PREDICTED LABEL")
plt.ylabel("TRUE LABEL")
plt.title('Confusion Matrix for MultinomialNB Classifier')

plt.show();
```

    Observations:
    Out of 98,805 instances, only 1757 false predictions.
    Accuracy of predictions is good.

### Evaluating accuracy of 3 types of NB Classifiers:

```python
#MultinomialNB
MNB = MultinomialNB()
MNB = MNB.fit(X_train, y_train)
```

```python
from sklearn.naive_bayes import BernoulliNB
BNB = BernoulliNB()
BNB = BNB.fit(X_train, y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB = GNB.fit(X_train, y_train)
```

```python
from sklearn import metrics
models = []
models.append(('Naive Bayes Classifier - Mutinomial', MNB))
models.append(('Naive Bayes Classifier - Bernoulli', BNB))
models.append(('Naive Bayes Classifier - Gaussian', GNB))
```

```python
#evaluation of train set
for i, j in models:
    scores = cross_val_score(j, X_train, y_train, cv=10)
    
    accuracy = metrics.accuracy_score(y_train, j.predict(X_train))
    
    confusion_matrix = metrics.confusion_matrix(y_train, j.predict(X_train))
    
    classification = metrics.classification_report(y_train, j.predict(X_train))
    print()
    print("Model Evaluation of {}.".format(i))
    print()
    print("Cross Validation Mean Score:" "\n", scores.mean())
    print()
    print("Model Accuracy:" "\n", accuracy)
    print()
    print("Confusion matrix:" "\n", confusion_matrix)
    print()
    print("Classification report:" "\n", classification) 
    print()
```

```python
#evaluation of test set
for i, j in models:
    accuracy = metrics.accuracy_score(y_test, j.predict(X_test))
    
    confusion_matrix = metrics.confusion_matrix(y_test, j.predict(X_test))
    
    classification = metrics.classification_report(y_test, j.predict(X_test))
    print()
    print("Model Test Evaluation of {}.".format(i))
    print()
    print("Model Accuracy:" "\n", accuracy)
    print()
    print("Confusion matrix:" "\n", confusion_matrix)
    print()
    print("Classification report:" "\n", classification) 
    print()        
```

#### Comments:
    All 3 classifiers have accuracy above 98% for test & train sets.
    Precision is more for intrusion predictions than for normal.

## More parametric evaluations on accuracy of Mutinomial NB Classifier:

```python
#ROC curve & AUC values:
from sklearn.metrics import precision_recall_curve, auc, roc_curve

#for alpha=0.01
#model fitting to train and test set
MNB = MultinomialNB(alpha=0.01)

MNB.fit(X_train, y_train)

train_fpr, train_tpr, thresholds = roc_curve(y_train, MNB.predict_proba(X_train)[:,1])

test_fpr, test_tpr, thresholds = roc_curve(y_test, MNB.predict_proba(X_test)[:,1])

plt.plot(train_fpr, train_tpr, label=" train AUC =" +str(auc(train_fpr, train_tpr)))

plt.plot(test_fpr, test_tpr, label=" test AUC =" +str(auc(test_fpr, test_tpr)))

plt.xlabel("hyperparameter:alpha")
plt.ylabel("AUC")
plt.title("Error Plots")
plt.legend()
plt.show();
```

    Observation:
    
    Train and test AUC curves are in line.
    So, there is no overfitting of model.
    Very accurate in predictions.

```python
#F1 scores of train and test set predictions
from sklearn.metrics import f1_score, recall_score

y_train_pred = MNB.predict(X_train)

y_test_pred = MNB.predict(X_test)

print('Train f1 score', f1_score(y_train, y_train_pred))

print('Test f1 score', f1_score(y_test, y_test_pred))

print("train recall score", recall_score(y_train, y_train_pred))

print("test recall score", recall_score(y_test, y_test_pred))
```

    Observations:
    
    F1 score is the inverse of average of precision and recall.
    It will give high value when both precision and recall is high.
    F1 scores indicate high accuracy of model in test and train sets.

### Hyperparameter tuning for different alpha values:

```python
#creating object of Multinomial Naive Bayes
multi_NB = MultinomialNB()

#hyper parameter initialization
parameters = {'alpha': [10**-4, 10**-3, 10**-2, 10**-1, 10**1, 10**2, 10**3,10**4]}
alpha_range = [10**-4, 10**-3, 10**-2, 10**-1, 10**1, 10**2, 10**3,10**4]

#putting the model in grid search cv to find the best hyper parameter
#classifier initialize; 10 fold cross validation
clf = GridSearchCV(multi_NB, parameters, cv=10, scoring='roc_auc', return_train_score=True)

#fitting X_train and y_train with the Multinomial Naive Bayes
clf.fit(X_train, y_train)

train_auc = clf.cv_results_['mean_train_score']

train_auc_std = clf.cv_results_['std_train_score']

cv_auc = clf.cv_results_['mean_test_score'] 

cv_auc_std= clf.cv_results_['std_test_score']

plt.plot(alpha_range, train_auc, label='Train AUC')

plt.gca().fill_between(alpha_range, train_auc - train_auc_std,train_auc + train_auc_std, alpha=0.2, color='darkblue')

plt.plot(alpha_range, cv_auc, label='CV AUC')

plt.gca().fill_between(alpha_range, cv_auc - cv_auc_std,cv_auc + cv_auc_std, alpha=0.2, color='darkorange')

plt.legend()
plt.xlabel("hyperparameter")
plt.xscale('log')
plt.ylabel("AUC")
plt.title("ERROR PLOTS")
plt.show();
```

### Conclusion:
    
    Classifier accuracy is above 99% when alpha parameter is hypertuned to
    different values ranging upto alpha=1000.
    
    Efficiency of model increased to 99% after crossvalidation & hypertuning.
    
    Further it is seen that accuracy scores decreases after alpha values of 1000 and above.
    
    Alpha values between 0.001 and 1000 can be selected as best parameters for the model.
