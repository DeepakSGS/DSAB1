# Iris Dataset - Assignment on Python for Data Science

```python
#import pandas, numpy, matplotlib and seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read dataset to Python environment

```python
#reading the iris dataset from iris.xls to variable iris_df
iris_df = pd.read_excel("iris.xls", sheet_name = 'Sheet1')
```

```python
#display top 5 rows with the function head()
iris_df.head()
print(iris_df.head())
```

# Display column list

```python
iris_df.columns
```

```python
#rename column list
iris_df.columns=['Sapel Length', 'Sapel Width', 'Petal Length', 'Petal Width', 'Classification']
```

```python
#print iris dataset
print(iris_df)
```

```python
#looking at data types in dataset
iris_df.dtypes
```

```python
#data description & info
iris_df.describe()
iris_df.info()
```

# Checking for Null values

```python
#checking for null values
iris_df.isnull()
print(iris_df.isnull())

#summarising the null values
iris_df.isnull().sum()
print("Summary of null values: \n" ,iris_df.isnull().sum())

# remove any empty lines
iris_df.dropna(how="all", inplace=True)
```

# Mean of each column

```python
#mean of a column
mean_data1 = iris_df["Sapel Length"].mean()
mean_data2 = iris_df["Sapel Width"].mean()
mean_data3 = iris_df["Petal Length"].mean()
mean_data4 = iris_df["Petal Width"].mean()
print("Mean value of Sapel Length:",mean_data1, "\nMean value of Sapel Width:", mean_data2,
      "\nMean value of Petal Length:",mean_data3, "\nMean value of Petal Width:", mean_data4)
```

# Mean value by Species

```python
#mean values by species
df_mean1 = iris_df[['Sapel Length','Sapel Width','Petal Length',
                    'Petal Width']][iris_df['Classification']=='Iris-setosa'].mean()
print("Iris-setosa mean of:\n", df_mean1)

df_mean2 = iris_df[['Sapel Length','Sapel Width','Petal Length',
                    'Petal Width']][iris_df['Classification']=='Iris-versicolor'].mean()
print("Iris-versicolor mean of:\n", df_mean2)

df_mean3 = iris_df[['Sapel Length','Sapel Width','Petal Length',
                    'Petal Width']][iris_df['Classification']=='Iris-virginica'].mean()
print("Iris-virginica mean of:\n", df_mean3)
```

# Perform meaningful visualizations using the dataset

#plot line graph

```python
#selecting only first 4 columns as they are the independent(X) variable
iris_x=iris_df.iloc[:,[0,1,2,3]]
iris_x.plot(kind='line')
plt.xticks(rotation = 0)
plt.show()
```

#plotting bar graph from mean values

```python
iris_x=iris_df.iloc[:,[0,1,2,3]].mean()
iris_x.plot(kind='bar')
plt.xlabel("Sapels & Petals")
plt.ylabel("Length & Width")
plt.title("Mean Values of Length & Width - Iris(Unclassified)")
plt.xticks(rotation =0)
plt.show()

iris = iris_df.groupby(['Classification']).mean()
irissl = iris.iloc[:,[0,1,2,3]]
irissl.plot(kind='bar')
plt.xlabel("Species")
plt.ylabel("Length & Width")
plt.title("Iris Species Distribution Chart - Length & Width")
plt.xticks(rotation =0)
plt.show()
```

#box plot

```python
sns.boxplot(x = 'Classification', y = 'Sapel Length', data = iris_df)
plt.title("Total Distribution", fontsize = 15)
plt.show()
```

#plot histogram

```python
x = iris_df["Sapel Length"]
plt.hist(x, bins = 20, rwidth = 0.9, color = "green")
plt.title("Sapel Length Distribution", fontsize = 15)
plt.xlabel("Sapel Length cm")
plt.ylabel("Count")
plt.xticks()
plt.yticks()
plt.show()

x = iris_df["Sapel Width"]
plt.hist(x, bins = 20, rwidth = 0.9, color = "yellow")
plt.title("Sapel Width Distribution", fontsize = 15)
plt.xlabel("Sapel Width cm")
plt.ylabel("Count")
plt.xticks()
plt.yticks()
plt.show()

x = iris_df["Petal Length"]
plt.hist(x, bins = 20, rwidth = 0.9, color = "red")
plt.title("Petal Length Distribution", fontsize = 15)
plt.xlabel("Petal Length cm")
plt.ylabel("Count")
plt.xticks()
plt.yticks()
plt.show()

x = iris_df["Petal Width"]
plt.hist(x, bins = 20, rwidth = 0.9,  color = "blue")
plt.title("Petal Width Distribution", fontsize = 15)
plt.xlabel("Petal Width cm")
plt.ylabel("Count")
plt.xticks()
plt.yticks()
plt.show()
```

#density estimation plot

```python
iris_df[["Sapel Length", "Sapel Width", "Petal Length", 
         "Petal Width"]].plot.kde(color = ["blue", "green", "orange", "red"])
plt.title("Density Estimation", fontsize = 15)
plt.show()
```

#scatter plot

```python
plt.scatter(x= "Petal Width", y = "Sapel Width", data = iris_df, s = 10, color ="blue")
plt.title("Scatter Plot - Width Distribution", fontsize = 15)
plt.xlabel("Petal Width cm")
plt.ylabel("Sapel Width")
plt.show()

plt.scatter(x= "Petal Length", y = "Sapel Length", data = iris_df, s = 10, color ="orange")
plt.title("Scatter Plot - Length Distribution", fontsize = 15)
plt.xlabel("Petal Length")
plt.ylabel("Sapel Length")
plt.show()
```

#violin plot

```python
plt.title("Violin Plot", fontsize = 10)
sns.violinplot(x = "Classification", y = "Sapel Length", data = iris_df)
plt.show()
```

#pair plot

```python
x = iris_df [['Sapel Length', 'Petal Length', 'Sapel Width', 'Petal Width']]
sns.pairplot(x)
plt.show()
```

#heatmap

```python
plt.figure(figsize = (10, 7))
x = x.corr()
sns.heatmap(x, annot = True)
plt.title("Heat Map", fontsize = 15)
plt.xticks(rotation = 0)
```
